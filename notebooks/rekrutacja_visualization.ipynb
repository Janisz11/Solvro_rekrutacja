```python
import pandas as pd
```

```python
cocktail_df =pd.read_json('cocktail_dataset.json')
```

```python
cocktail_df.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns





sns.set(style="whitegrid")


category_counts = cocktail_df['category'].value_counts()

plt.figure(figsize=(10, 6))
sns.barplot(x=category_counts.index, y=category_counts.values, palette="Blues_d")
plt.title("Number of Cocktails per Category")
plt.xlabel("Category")
plt.ylabel("Number of Cocktails")
plt.xticks(rotation=45)
plt.show()


glass_counts = cocktail_df['glass'].value_counts()

plt.figure(figsize=(10, 6))
sns.barplot(x=glass_counts.index, y=glass_counts.values, palette="Oranges_d")
plt.title("Types of Glasses Used")
plt.xlabel("Glass Type")
plt.ylabel("Frequency")
plt.xticks(rotation=45)
plt.show()


alcoholic_counts = cocktail_df['alcoholic'].map({1: 'Alcoholic', 0: 'Non-Alcoholic'}).value_counts()

plt.figure(figsize=(6, 6))
plt.pie(alcoholic_counts, labels=alcoholic_counts.index, autopct='%1.1f%%', startangle=140, colors=sns.color_palette("Set2"))
plt.title("Alcoholic vs Non-Alcoholic Cocktails")
plt.ylabel("")
plt.show()

cocktail_df['ingredient_count'] = cocktail_df['ingredients'].apply(len)
ingredient_counts = cocktail_df['ingredient_count'].value_counts()

plt.figure(figsize=(10, 6))
sns.barplot(x=ingredient_counts.index, y=ingredient_counts.values, palette="Greens_d")
plt.title("Number of Ingredients per Cocktail")
plt.xlabel("Number of Ingredients")
plt.ylabel("Number of Cocktails")
plt.show()
```
